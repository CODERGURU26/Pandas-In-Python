# Group By & Aggregate Function In Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\gs602\OneDrive\Documents\Group By & Aggregations In Pandas\Flavors.csv")
df
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').sum(numeric_only = True)
```

```python
df.groupby('Base Flavor').mean(numeric_only=True)
```

```python
df.groupby(['Base Flavor' , 'Liked']).agg({'Flavor Rating' : ['mean' , 'min' , 'max' , 'count' , 'sum']})
```
