# Visualization In Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\gs602\OneDrive\Documents\Visualization In Pandas\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('tableau-colorblind10')
```

# Line Graph

```python
df.plot(kind = 'line' , 
        title ='Icecream Ratings' ,
       xlabel = 'Daily Rating',
       ylabel = 'Ratings')
```

# Bar Graph

```python
df.plot(kind = 'bar' , 
        title ='Icecream Ratings' ,
       xlabel = 'Daily Rating',
       ylabel = 'Ratings')
```

```python
df.plot(kind = 'bar' ,
        stacked = True,
        title ='Icecream Ratings' ,
       xlabel = 'Daily Rating',
       ylabel = 'Ratings')
```

```python
df.plot.barh(
        title ='Icecream Ratings' ,
       xlabel = 'Daily Rating',
       ylabel = 'Ratings')
```

```python
df.plot.barh( stacked = True ,
        title ='Icecream Ratings' ,
       xlabel = 'Daily Rating',
       ylabel = 'Ratings')
```

# Scatter Plot

```python
df.plot.scatter(x = 'Texture Rating' , y = 'Overall Rating' , s  = 100 , c = 'red')
```

# Histogram

```python
df.plot.hist(bins = 10)
```

# Box Plot

```python
df.plot.box()
```

# Area Plot

```python
df.plot.area()
```

# Pie Chart

```python
df.plot.pie(y='Flavor Rating')
```
